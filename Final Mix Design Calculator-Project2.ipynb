```python
#imports packages

import pandas as pd
import numpy as np
```

```python
#defines global constants for conformity across functions
cubic_yard_ft3 = 27
weight_water = 62.4
```

```python
#Defines a Function for calculating the Weight of Water
def water_weight(weight_cement, weight_flyash, weight_silica, weight_other, WC_ratio) :
    
    total_cement = weight_cement + weight_flyash + weight_silica + weight_other
    water_weight =  total_cement*WC_ratio
   
    return water_weight

#Defines a Function for calculating the Cement Volume
def cement_volume(weight_cement, specific_gravity):

    specific_water = specific_gravity*weight_water

    cement_volume = weight_cement/specific_water

    return cement_volume

#Defines a Function for calculating the Fly Ash Volume
def flyash_volume(weight_ash, specific_gravity2):

    specific_water2 = specific_gravity2*weight_water

    flyash_volume = weight_ash/specific_water2

    return flyash_volume

#Defines a Function for calculating the Silica Fume Volume
def silica_volume(weight_fume, specific_gravity3):

    specific_water3 = specific_gravity3*weight_water

    silica_volume = weight_fume/specific_water3

    return silica_volume

#Defines a Function for calculating the Other SCM Volume
def other_volume(weight_SCM, specific_gravity4):
    
    specific_water4 = specific_gravity4*weight_water
    
    other_volume = weight_SCM/specific_water4

    return other_volume

#Defines a Function for calculating the Air Volume
def air_volume(target_air_percent,) :

    content_value = target_air_percent / 100

    air_volume = content_value * cubic_yard_ft3

    return air_volume

#Defines a Function for calculating the Water Volume
def water_volume(Q) :
    water_volume = Q / weight_water
    return water_volume

#Defines a Function for calculating the Total Aggregate Volume
def AggregateVolume(volume_R,volume_S,volume_T,volume_U,volume_V,volume_W) :

    X = cubic_yard_ft3 - volume_R - volume_S - volume_T - volume_U - volume_V - volume_W
    return X

#Defines a Function for calculating the Fine Aggregate Weight
def fine_agg_weight(specific_gravity_fine, X, target_percent_fine):

    calculate_fine = weight_water * (target_percent_fine / 100) * specific_gravity_fine * X
    
    return calculate_fine

#Defines a Function for calculating the Coarse Aggregate Weight
def coarse_agg_weight(specific_gravity_coarse, X, target_percent_coarse):

    calculate_coarse = weight_water * (target_percent_coarse / 100) * specific_gravity_coarse * X
    
    return calculate_coarse

#Defines a Function for calculating the Other Aggregate Weight
def other_agg_weight(specific_gravity_other, X, target_percent_other):

    calculate_other= weight_water * (target_percent_other / 100) * specific_gravity_other * X
    
    return calculate_other
```

```python
#Asks user For project Meta Data
project_no = int(input("Enter project number: "))
concrete_class = input("Enter class of concrete: ") 

#Asks User for Cementuious Weights
input_cement = float(input("Enter cement weight  (lb per cubic yard): "))
input_ash = float(input("Enter fly ash weight weight  (lb per cubic yard): "))
input_silica = float(input("Enter silica fume weight  (lb per cubic yard): "))
input_other = float(input("Enter other SCM weight  (lb per cubic yard): "))

#Asks Users for W/C ratio and air percent
input_WC_rato = float(input("Enter the WC ratio : "))
input_target_air_percent = float(input("Enter the target air percent: "))

#Asks Users for Aggregete percentages
input_fine = float(input("Enter fine aggregate target percent  (lb per cubic yard): "))
input_coarse = float(input("Enter coarse aggregate target percent (lb per cubic yard): "))
input_other_agg = float(input("Enter other aggregate target percent (lb per cubic yard): "))

#Asks Users for specific gravties for materials
sg_cement = float(input("Enter specific gravity of the cement: ")) 
sg_ash = float(input("Enter specific gravity of the fly ash: ")) 
sg_silica = float(input("Enter specific gravity of the silica fume: ")) 
sg_other = float(input("Enter specific gravity of the other SCM: ")) 
sg_fine = float(input("Enter specific gravity of the fine aggregate: ")) 
sg_coarse = float(input("Enter specific gravity of the coarse aggregate: ")) 
sg_otheragg = float(input("Enter specific gravity of the other aggregate: "))

#Defines the inputs as our variables used in the defined functions
weight_cement = input_cement
weight_ash = input_ash
weight_fume = input_silica
weight_SCM = input_other
WC_ratio = input_WC_rato
specific_gravity = sg_cement
specific_gravity2 = sg_ash
specific_gravity3 = sg_silica
specific_gravity4 = sg_other
target_air_percent = input_target_air_percent
specific_gravity_fine = sg_fine
specific_gravity_coarse = sg_coarse
specific_gravity_other = sg_otheragg
target_percent_fine = input_fine
target_percent_coarse = input_coarse
target_percent_other = input_other_agg

#Assigns a Variable to the calculated water weight
Q = water_weight(weight_cement,weight_ash, weight_fume, weight_SCM, WC_ratio)

#Assigns Variables to the calculated Volumes
volume_R = cement_volume(weight_cement, specific_gravity)
volume_S = flyash_volume(weight_ash, specific_gravity2)
volume_T=  silica_volume(weight_fume, specific_gravity3)
volume_U = other_volume(weight_SCM, specific_gravity4)
volume_V = air_volume(target_air_percent)
volume_W = water_volume(Q)

#Assigns a variable to the calculated total aggregate volume
X= AggregateVolume( volume_R, volume_S, volume_T, volume_U, volume_V, volume_W)

#Assigns Variables to the calculated Aggregate Weights
fine_agg = fine_agg_weight(specific_gravity_fine, X, target_percent_fine)
coarse_agg = coarse_agg_weight(specific_gravity_coarse, X, target_percent_coarse)
other_agg = other_agg_weight(specific_gravity_other, X, target_percent_other)

#Prints out the Final Mix Design Output
print("\n---------------------------------------------")
print(f" Cement Volume(R): {volume_R:8.2f} ft^3")
print(f" Fly Ash Volume Volume(S): {volume_S:8.2f} ft^3")
print(f" Silica Fume Volume(T): {volume_T:8.2f} ft^3")
print(f" Other SCM Volume(U): {volume_U:8.2f} ft^3")
print(f" Air Volume(V): {volume_V:8.2f} ft^3")
print(f" Water Volume(W): {volume_W:8.2f} ft^3")
print(f"Total Aggregate Volume (X): {X:8.2f} ft^3")
print(f"Weight of Water (Q):    {Q:8.0f} lbs")
print("---------------------------------------------")
print("\n---------------------------------------------")
print(" NDOT Concrete Mix Design – Weight Summary")
print("         (1 Cubic Yard of Concrete)")

print("---------------------------------------------")
print("\n---------------------------------------------")
print(f"Project Number: {project_no}")
print(f"Cement class: {concrete_class}")
print("---------------------------------------------")
print("\n---------------------------------------------")
print(f"Cement (A):     {input_cement:8.1f} lbs")
print(f"Fly Ash (B):    {input_ash:8.1f} lbs")
print(f"Silica Fume (C):    {input_silica:8.1f} lbs")
print(f"Other SCM (D):    {input_other:8.1f} lbs")
print("---------------------------------------------")
print("\n---------------------------------------------")
print(f"Fine Aggregate (Y):    {fine_agg:8.0f} lbs")
print(f"Coarse Aggregate (Z):    {coarse_agg:8.0f} lbs")
print(f"Other Aggregate (AA):    {other_agg:8.0f} lbs")
print("---------------------------------------------")
print("\n---------------------------------------------")
print(f"Water Weight (Q):    {Q:8.0f} lbs")
print("---------------------------------------------")
print("Calculations complete")
print("---------------------------------------------")

```
